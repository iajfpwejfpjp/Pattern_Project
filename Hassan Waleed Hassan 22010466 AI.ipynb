```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

```python
column_names = ['age', 'workclass', 'fnlgwt', 'education', 'education-num', 'marital-status', 'occupation','relationship',
                'race', 'sex', 'capital-gain', 'capital-loss','hours-per-week', 'native-country', 'income']
train = pd.read_csv("D:/adult.data")
train.columns=column_names
Test = pd.read_csv("D:/adult.test", skiprows=1)
Test.columns = column_names
train.dropna(inplace=True)
Test.dropna(inplace=True)
df=pd.concat([train, Test], axis=0)
df_encoded= pd.get_dummies(df)
midpoint = len(df_encoded) // 2
df1 = df_encoded.iloc[:midpoint]
df2 = df_encoded.iloc[midpoint:].reindex(columns=df1.columns)
df2


```

```python
X_train = df1.drop('income_ >50K', axis=1)
y_train = df1['income_ >50K']
Model = GaussianNB()
Model.fit(X_train, y_train)
```

```python
X_test = df2.drop('income_ >50K', axis=1)
y_test = df2['income_ >50K']
predictions = Model.predict(X_test)
```

```python
tn, fp, fn, tp = confusion_matrix(y_train, predictions).ravel()
sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)

print(f"Sensitivity: {sensitivity}")
print(f"Specificity:, {specificity}")
```

```python
posterior_probabilities = Model.predict_proba(X_test)
Over50k = posterior_probabilities[:, 1]
print(f"Posterior Probablity = {Over50k}")
 
```
